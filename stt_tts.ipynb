```python
from google.colab import files
uploaded = files.upload()

audio_file = list(uploaded.keys())[0]
print("Uploaded:", audio_file)
```

```python
from pydub import AudioSegment

sound = AudioSegment.from_file(audio_file)
sound = sound.set_frame_rate(16000).set_channels(1)
sound.export("clean.wav", format="wav")

print("Audio cleaned successfully")
```

```python
import speech_recognition as sr

recognizer = sr.Recognizer()

with sr.AudioFile("clean.wav") as source:
    audio = recognizer.record(source)

try:
    text = recognizer.recognize_google(audio)
    print("Recognized Text:", text)
except:
    print("Speech not clear")
```

```python
from langdetect import detect

language = detect(text)
print("Detected Language:", language)
```

```python
def multilingual_ai(text, language):

    text = text.lower()

    if "stress" in text or "तनाव" in text:
        if language == "hi":
            return "चिंता मत करो। सब ठीक हो जाएगा।"
        elif language == "bn":
            return "চিন্তা করবেন না। সব ঠিক হয়ে যাবে।"
        else:
            return "Don't worry. Everything will be okay."

    elif "happy" in text or "खुश" in text:
        if language == "hi":
            return "यह सुनकर अच्छा लगा!"
        elif language == "bn":
            return "এটা শুনে ভালো লাগলো!"
        else:
            return "That’s wonderful to hear!"

    else:
        return "I understand you. Tell me more."
```

```python
response = multilingual_ai(text, language)
print("AI Response:", response)
```

```python
import asyncio
import edge_tts
from IPython.display import Audio

async def speak(text, language):

    if language == "hi":
        voice = "hi-IN-MadhurNeural"
    elif language == "bn":
        voice = "bn-IN-BashkarNeural"
    else:
        voice = "en-IN-PrabhatNeural"

    communicate = edge_tts.Communicate(text, voice)
    await communicate.save("response.mp3")

await speak(response, language)

Audio("response.mp3")
```
